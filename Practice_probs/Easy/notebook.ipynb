## 217, contains duplicates 



### Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.



```python
from typing import List
class ContainsDuplicate:
    def __init__(self, nums:List[int]) -> bool:

        ## using the set method to simply remove the duplicates
        contains = set()
        for i in nums: 
            if i in contains: 
                return True
            contains.add(i)
        
        return False
```

## Given an integer array nums and an integer k, return true if there are two distinct indices i and j in the array such that nums[i] == nums[j] and abs(i - j) <= k.

 


```python
class ContainsDuplicate2:
    def __init__(self,nums:List[int], k:int) -> bool:

        ## we are using the sliding window concept here to solve this problem 
        ### this problem has two variations, one by using the duplicates function to store the duplicates 
        ### sliding window to check whether the contains has the duplicates for the i,j with the stored capacity of k 


        contains = set(nums)
        for i,x in enumerate(nums):
            if x in contains: 
                return True 
            
            if i >= k: 
                contains.remove(nums[i-k]) 
            
        
        return False

        
        
```

3637. Trionic Array 


You are given an integer array nums of length n.

An array is trionic if there exist indices 0 < p < q < n − 1 such that:

nums[0...p] is strictly increasing,
nums[p...q] is strictly decreasing,
nums[q...n − 1] is strictly increasing.
Return true if nums is trionic, otherwise return false.



```python
class TrionicArray:
    def __init__(self,nums:List[int])  -> bool: 

        ## basically trionic array is the array where the order of the pattern in the array should be increasing, decreasing and strictly increasing
        ## in that case, we have to have the array with more than 3 elements 

        if len(nums) < 3: 
            return False 
        
        ## now that the array may have more than 3 elements and should follow this pattern so we can confirm that it is a trionic array 
        ## lets start from the first element of the array 

        i = 0 

        while i + 1 < len(nums)  and nums[i] < nums[i+1]: ## this checks if the current array element is less than the length of the array 
            i += 1 
        
        #now this condition checks for the increasing part 
        if i == 0 or i == len(nums) -1: 
            return False    ## since the array is trionic and it needs atleast 3 elements and further parts for the decreasing and increasing part as well 
        
        while i+1 < len(nums) and nums[i] > nums[i+1]: 
            i += 1 
        
        if i == len(nums) -1: 
            return False 
        
        while i+1 < len(nums) and nums[i] < nums[i+1]: 
            i += 1 
        
        return i == len(nums) -1    
         
        

                    


        
```

## 83. Remove duplicates from sorted list 
Given the head of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list sorted as well.



```python
class ListNode:
    def __init__(self,val=0,next=None):
        self.val = val 
        self.next = next 

class RemoveDuplicatesFromSortedList:
    def __init__(self,head:ListNode) -> ListNode:

        # with the oops logic of objects and references 
        dummy = head  # here head is the object instance of the listnode class and the dummmy is the reference of the object 

        while dummy and dummy.next:    ## as long as there is an element and the next element present in the dummy object this loop works   
            if dummy.val == dummy.next.val: #if the current value that is the first value of the object 
                dummy.next = dummy.next.next  # if the next element matches with the first instance, move to the next element ( basically skip)
            else: 
                dummy = dummy.next # if there is no duplicates found in the list , this will just loop through the next elements 
        
        return head # it will now return the final listnode object with all the duplicates removal logic 

        
            
```

## 141. Linked list Cycle 

Given head, the head of a linked list, determine if the linked list has a cycle in it.

There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.

Return true if there is a cycle in the linked list. Otherwise, return false.

 

  

```python

class ListNode:
    def __init__(self,x):
        self.val = x 
        self.next = None 
    

class LinkedlistCycle:
    def __init__(self,head:ListNode) -> bool:

        # lets use Flyod's cycle 
        single_Step = head 
        double_Step = head 


        while single_Step and double_Step.next:
            single_Step_1 = single_Step.next
            double_Step_1 = double_Step.next.next 

            if single_Step_1 == double_Step_1:
                return True 
            
        return False








```

## single number
Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.

You must implement a solution with a linear runtime complexity and use only constant extra space.

 

Example 1:

Input: nums = [2,2,1]

Output: 1

Example 2:

Input: nums = [4,1,2,1,2]

Output: 4

Example 3:

Input: nums = [1]

Output: 1





```python
class singlenumber:
    def __init__(self,nums:List[int]) -> int:

        #given the space constraints we cant use any other additional data structures here , so 

        single_num = 0 

        for i in nums:
            single_num ^= i 
        
        return single_num 
```

Given the heads of two singly linked-lists headA and headB, return the node at which the two lists intersect. If the two linked lists have no intersection at all, return null.

For example, the following two linked lists begin to intersect at node c1:

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class LinkedList:
    def __init__(self,val):
        self.val = val 
        self.next = None 

class Intersection:
    def __init__(self,headA:LinkedList,headB:LinkedList)->ListNode:

        ## USE THE OBJECT REFERENCES TO traverse through the Objects
        list1 = headA
        list2 = headB 

        while list1 is not list2:
            if list1 is not None:
                list1 = list1.next
            else:
                list1 = headB

            if list2 is not None: 
                list2 = list2.next 
            else:
                list2 = headA
                
        return list1
```

## Path Sum 


```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Pathsum:
    def __init__(self, root:TreeNode, targetSum:int) -> bool:

        if root is None:
            return false 

        if not root.left or not root.right:
            return targetSum == root.val 
        

        targetSum -= root.val 

        return (self.Pathsum(root.left,targetSum) or self.PathSum(root.right,targetSum))
```
