### duplicate zeros


### Input: arr = [1,0,2,3,0,4,5,0]
### Output: [1,0,0,2,3,0,0,4]

```python
class DuplicateZero:
    def __init__(self, arr:List[int]) -> None:

        trace = [] 


        for i in arr:
            trace.append(i)
            if i == 0:
                trace.append(0)
            
        ## since the constraint enforce the length of the array 
        for i in range(len(arr)):
            arr[i] = trace[i]
            

## this approach has a space constraint  O(n)

```

## optimal approach 

```python
from re import L


class DuplicateZero:
    def __init__(self, arr:List[int]) -> None:
        
        n = len(arr) 
        # using two pointer approach 
        i = n - 1  # last element of the arr 
        zeros = arr.count(0) # counts the number of zeros 
        j = i + zeros # the auxillary fill 

        while i < j: # this runs till true 
            if j < n: # auxillary fill should be constrained to the list length 
                arr[j] = arr[i] # swap 

            if arr[i] == 0: # zero detection
                j -= 1  # reduce the auxillary space 
                if j < n: # edge case ,  prefilling zero 
                    arr[j] == 0 
            i -= 1  # pointer move 
            j -= 1 
```
