### duplicate zeros


### Input: arr = [1,0,2,3,0,4,5,0]
### Output: [1,0,0,2,3,0,0,4]

```python
class DuplicateZero:
    def __init__(self, arr:List[int]) -> None:

        trace = [] 


        for i in arr:
            trace.append(i)
            if i == 0:
                trace.append(0)
            
        ## since the constraint enforce the length of the array 
        for i in range(len(arr)):
            arr[i] = trace[i]
            

## this approach has a space constraint  O(n)

```

## optimal approach 

```python
from re import L


class DuplicateZero:
    def __init__(self, arr:List[int]) -> None:
        
        n = len(arr) 
        # using two pointer approach 
        i = n - 1  # last element of the arr 
        zeros = arr.count(0) # counts the number of zeros 
        j = i + zeros # the auxillary fill 

        while i < j: # this runs till true 
            if j < n: # auxillary fill should be constrained to the list length 
                arr[j] = arr[i] # swap 

            if arr[i] == 0: # zero detection
                j -= 1  # reduce the auxillary space 
                if j < n: # edge case ,  prefilling zero 
                    arr[j] == 0 
            i -= 1  # pointer move 
            j -= 1 
```

## transpose of a matrix

```python
class Transpose:
    def __init__(self, matrix : List[List[int]]) -> List[List[int]] :

        row  = len(matrix)
        col = len(matrix[0])

        result = [[0 for i in range(row)] for j in range(col)]

        for i in range(row):
            for j in range(col):
                result[j][i] = matrix[i][j]
        
        return  result
```

## reverse a string 

```python
class ReverseString:
    def __init__(self, s:List[str]) -> None:

        # using two pointer approach 
        l , r = 0 , len(s)-1 

        while l < r:
            s[l], s[r] = s[r], s[l]
            l += 1 
            r -= 1 
```

### issubsequence

```python
class IsSubsequence:
    def __init__(self, s:str, t: str) -> bool:
        if not s:
            return True 
        
        # in this approach we are just comparing the length of the comparing string s with count 
        count = 0 

        for i in t:
            if count < len(s) and i == s[count]:
                count += 1 
        
        return count == len(s) # this basically returns true if matches else False
```

## Reverse vowels of a string 

Given a string s, reverse only all the vowels in the string and return it.

The vowels are 'a', 'e', 'i', 'o', and 'u', and they can appear in both lower and upper cases, more than once.



```python
class Reverse:
    def __init__(self, s : str) -> str:

        v = set("aeiouAEIOU")

        e =  [c for c in s if c in v] ## extracting the vowels out of the string 
        r = e.reverse()  ## exactly reversed the string 


        output = [] 
        start = 0 
        
        for i in s:
            if i in v:
                output.append(r[start])
                start += 1 
            else:
                output.append(i)
        
        return "".join(output)
```

## check if  n and its double exist 

```python
class Check:
    def __init__(self, arr:List[int]) -> bool :

        seen  = set()   # we will first hold this for now

        for nums in arr:
            # only two variants are possible if the number is either even or odd , if even :   * 2 and / 2 should exist
            if 2 * nums or (nums%2 == 0 and nums // 2 ==0 in seen):
                return True
            seen.add(nums)

        return False 
```

## Minimum Distance Between Three Equal Elements I

You are given an integer array nums.

A tuple (i, j, k) of 3 distinct indices is good if nums[i] == nums[j] == nums[k].

The distance of a good tuple is abs(i - j) + abs(j - k) + abs(k - i), where abs(x) denotes the absolute value of x.

Return an integer denoting the minimum possible distance of a good tuple. If no good tuples exist, return -1.

Example 1:

Input: nums = [1,2,1,1,3]

Output: 6

Explanation:

The minimum distance is achieved by the good tuple (0, 2, 3).

(0, 2, 3) is a good tuple because nums[0] == nums[2] == nums[3] == 1. Its distance is abs(0 - 2) + abs(2 - 3) + abs(3 - 0) = 2 + 1 + 3 = 6.

Example 2:

Input: nums = [1,1,2,3,2,1,2]

Output: 8

Explanation:

The minimum distance is achieved by the good tuple (2, 4, 6).

(2, 4, 6) is a good tuple because nums[2] == nums[4] == nums[6] == 2. Its distance is abs(2 - 4) + abs(4 - 6) + abs(6 - 2) = 2 + 2 + 4 = 8.

Example 3:

Input: nums = [1]

Output: -1

Explanation:

There are no good tuples. Therefore, the answer is -1.

```python
class Mindist:
    def __init__(self, nums:List[int]) -> int:
        n = len(nums) # get the  length 
        min_d = float("inf") # just a placeholder 

        map = defaultdict(list) # { key : [value]}

        for i,x in enumerate(nums):
            map[x].append(i) # {1 : [1,3,4]} 

            if len(map[x]) >= 3:
                distance = 2 * (map[x][-1] - map[x][-3]) # 2 * (c - a)
                min_d  = min(min_d, distance)
        

        return min_d if min_d != float('inf')  else -1 
```

## Reverse String II

Given a string s and an integer k, reverse the first k characters for every 2k characters counting from the start of the string.

If there are fewer than k characters left, reverse all of them. If there are less than 2k but greater than or equal to k characters, then reverse the first k characters and leave the other as original.

 

Example 1:

Input: s = "abcdefg", k = 2
Output: "bacdfeg"
Example 2:

Input: s = "abcd", k = 2
Output: "bacd"

```python
class reverseString:
    def __init__(self, s : str, k : int) -> str:

        ## since string are immutable and we need to convert them to mutable to avoid additional space complexity 
        str_l = list(s)
        n = len(str_l)

        if n <= k:
            return "".join(str_l[::-1])
        
        ## move the string by parts of 2*k
        for i in range(0,len(s), 2 * k):
            str_l[i : i + k] = str_l[i:i+k][::-1] 
        
        return "".join(str_l) 
```
